```python
import logging
from ipynb.fs.defs.func_prediction import is_user_fat_3
import time


from telegram import ReplyKeyboardMarkup, ReplyKeyboardRemove, Update
from telegram.ext import (
    Updater,
    CommandHandler,
    MessageHandler,
    Filters,
    ConversationHandler,
    CallbackContext,
)

# Enable logging
logging.basicConfig(
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO
)

logger = logging.getLogger(__name__)

#create states
(SEX, AGE, HEIGHT, WEIGHT, FAMILY_OVERW, FAVC, FCVC, NCP,
 CAEC, SMOKE, CH2O, SCC, FAF, TUE, CALC, MTRANS) = range(16)

# list for the user's answers
list_ = []


def start(update: Update, _: CallbackContext) -> int:
    reply_keyboard = [['Female', 'Male']]

    update.message.reply_text('Hello!!! It is obesity_classification_bot. I am going to ask you some questions. '
                              'Based on your answers, I will classify you to one or another weight type.\n\n'
                              'Let\'s start!\n'
                              'What is your Sex?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return SEX


def sex(update: Update, _: CallbackContext) -> int:
    user = update.message.from_user
    if update.message.text == 'Female':
        list_.append('0')
    else:
        list_.append('1')
        
    logger.info("Sex of %s: %s", user.first_name, update.message.text)
    update.message.reply_text('What is your Age?\n'
                              '• Numeric value')

    return AGE


def age(update: Update, _: CallbackContext) -> int:
    user = update.message.from_user
    list_.append(update.message.text)
    print(list_)
    logger.info("Age of %s: %s", user.first_name, update.message.text)
    update.message.reply_text('What is your Height?\n'
                              '• Numeric value in centimeters')

    return HEIGHT


def height(update: Update, _: CallbackContext) -> int:
    user = update.message.from_user
    list_.append(update.message.text)

    logger.info("Height of %s: %s", user.first_name, update.message.text)
    update.message.reply_text('What is your Weight?\n'
                              '• Numeric value in kilograms')

    return WEIGHT


def weight(update: Update, _: CallbackContext) -> int:

    list_.append(update.message.text)

    reply_keyboard = [['No', 'Yes']]
    update.message.reply_text('Do you have overweight people in your family?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return FAMILY_OVERW


def family_overweight(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'No':
        list_.append('0')
    else:
        list_.append('1')

    reply_keyboard = [['No', 'Yes']]
    update.message.reply_text('Do you eat high caloric food frequently?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return FAVC


def favc(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'No':
        list_.append('0')
    else:
        list_.append('1')

    reply_keyboard = [['Never', 'Sometimes', 'Always']]
    update.message.reply_text('Do you often eat vegetables?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return FCVC


def fcvc(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'Never':
        list_.append('1')
    elif update.message.text == 'Sometimes':
        list_.append('2')
    else:
        list_.append('3')

    reply_keyboard = [['One', 'Two', 'Three', 'More than three']]
    update.message.reply_text('How many main meals do you have daily?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return NCP


def ncp(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'One':
        list_.append('1')
    elif update.message.text == 'Two':
        list_.append('2')
    elif update.message.text == 'Three':
        list_.append('3')
    else:
        list_.append('4')

    reply_keyboard = [['No', 'Sometimes', 'Frequently', 'Always']]
    update.message.reply_text('Do you eat any food between meals?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return CAEC


def caec(update: Update, _: CallbackContext) -> int:

    if update.message.text == 'No':
        list_.append('no')
    else:
        list_.append(update.message.text)

    reply_keyboard = [['No', 'Yes']]
    update.message.reply_text('Do you smoke?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return SMOKE


def smoke(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'No':
        list_.append('0')
    else:
        list_.append('1')
        
    reply_keyboard = [['Less than a liter',
                       'Between 1 and 2 L', 'More than 2 L']]
    update.message.reply_text('How much liters of water do you drink daily?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return CH2O


def ch2o(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'Less than a liter':
        list_.append('1')
    elif update.message.text == 'Between 1 and 2 L':
        list_.append('2')
    else:
        list_.append('3')

    reply_keyboard = [['No', 'Yes']]
    update.message.reply_text('Do you monitor the calories you eat daily?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return SCC


def scc(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'No':
        list_.append('0')
    else:
        list_.append('1')

    reply_keyboard = [
        ['I do not have', '1 or 2 days', '3 or 4 days', '5 and more']]
    update.message.reply_text('How often do you have physical activity in a week?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return FAF


def faf(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'I do not have':
        list_.append('0')
    elif update.message.text == '1 or 2 days':
        list_.append('1')
    elif update.message.text == '2 or 4 days':
        list_.append('2')
    else:
        list_.append('3')

    reply_keyboard = [['0–2 hours', '3–5 hours', 'More than 5 hours']]
    update.message.reply_text('How much time do you use technological devices such as smartphones,'
                              'television, computer and others?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return TUE


def tue(update: Update, _: CallbackContext) -> int:
    if update.message.text == '0–2 hours':
        list_.append('0')
    elif update.message.text == '3–5 hours':
        list_.append('1')
    else:
        list_.append('2')

    reply_keyboard = [['I do not drink', 'Sometimes', 'Frequently', 'Always']]
    update.message.reply_text('How often do you drink alcohol?',
                              reply_markup=ReplyKeyboardMarkup(
                                  reply_keyboard, one_time_keyboard=True)
                              )

    return CALC


def calc(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'I do not drink':
        list_.append('no')
    else:
        list_.append(update.message.text)

    reply_keyboard = [['Automobile', 'Motorbike',
                       'Bike', 'Public Transportation', 'Walking']]
    update.message.reply_text('What kind of transportation do you usually use?',
                              reply_markup=ReplyKeyboardMarkup(reply_keyboard, one_time_keyboard=True))

    return MTRANS


def mtrans(update: Update, _: CallbackContext) -> int:
    if update.message.text == 'Public Transportation':
        list_.append('Public_Transportation')
    else:
        list_.append(update.message.text)

    if len(list_) == 16:
        answer = is_user_fat_3(list_)
    else:
        answer = is_user_fat_3(list_[-16:])
        
    update.message.reply_text('Analyzing :) Wait, please')
    time.sleep(5)

    if int(answer) == 6.0:
        update.message.reply_text('With a probability of 97% you have  "Obesity Type III".\n'
                                  'I think you should visit a doctor.')
    elif int(answer) == 5.0:
        update.message.reply_text('With a probability of 97% you have  "Obesity Type II".\n'
                                  'I think you should visit a doctor.')
    elif int(answer) == 4.0:
        update.message.reply_text('With a probability of 97% you have  "Obesity Type I".\n'
                                  'I think you should visit a doctor.')
    elif int(answer) == 3.0:
        update.message.reply_text('With a probability of 97% you have  "Overweight Level II".\n'
                                  'Be careful. Walk more and do sports.')
    elif int(answer) == 2.0:
        update.message.reply_text('With a probability of 97% you have  "Overweight Level I".\n'
                                  'Be careful. Walk more and do sports.')
    elif int(answer) == 1.0:
        update.message.reply_text('With a probability of 97% you have  "Normal Weight"!\n'
                                  'Congratulations!')
    else:
        update.message.reply_text('With a probability of 97% you have  "Insufficient Weight"\n'
                                  'Eat more and stop dieting))')

    return ConversationHandler.END


def cancel(update: Update, _: CallbackContext) -> int:
    user = update.message.from_user
    logger.info("User %s canceled the conversation.", user.first_name)
    update.message.reply_text(
        'Bye! I hope we can talk again some day.', reply_markup=ReplyKeyboardRemove())

    return ConversationHandler.END


def main() -> None:
    # Create the Updater and pass it your bot's token.
    updater = Updater("Here is your token")

    # Get the dispatcher to register handlers
    dispatcher = updater.dispatcher

    # Add conversation handler with the states
    conv_handler = ConversationHandler(entry_points=[CommandHandler('start', start)],
                                       states={
        SEX: [MessageHandler(Filters.regex('^(Female|Male)$'), sex)],
        AGE: [MessageHandler(Filters.text, age)],
        HEIGHT: [MessageHandler(Filters.text, height)],
        WEIGHT: [MessageHandler(Filters.text, weight)],
        FAMILY_OVERW: [MessageHandler(Filters.regex('^(No|Yes)$'), family_overweight)],
        FAVC: [MessageHandler(Filters.regex('^(No|Yes)$'), favc)],
        FCVC: [MessageHandler(Filters.regex('^(Never|Sometimes|Always)$'), fcvc)],
        NCP: [MessageHandler(Filters.regex('^(One|Two|Three|More than three)$'), ncp)],
        CAEC: [MessageHandler(Filters.regex('^(No|Sometimes|Frequently|Always)$'), caec)],
        SMOKE: [MessageHandler(Filters.regex('^(No|Yes)$'), smoke)],
        CH2O: [MessageHandler(Filters.regex('^(Less than a liter|Between 1 and 2 L|More than 2 L)$'), ch2o)],
        SCC: [MessageHandler(Filters.regex('^(No|Yes)$'), scc)],
        FAF: [MessageHandler(Filters.regex('^(I do not have|1 or 2 days|3 or 4 days|5 and more)$'), faf)],
        TUE: [MessageHandler(Filters.regex('^(0–2 hours|3–5 hours|More than 5 hours)$'), tue)],
        CALC: [MessageHandler(Filters.regex('^(I do not drink|Sometimes|Frequently|Always)$'), calc)],
        MTRANS: [MessageHandler(Filters.regex('^(Automobile|Motorbike|Bike|Public Transportation|Walking)$'), mtrans)],
    },
        fallbacks=[CommandHandler('cancel', cancel)],
    )

    dispatcher.add_handler(conv_handler)

    # Start the Bot
    updater.start_polling()

    updater.idle()


if __name__ == '__main__':
    main()
```
